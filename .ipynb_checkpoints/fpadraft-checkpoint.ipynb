```python
from pathlib import Path
import glob

import numpy as np
import matplotlib.pyplot as plt
import pyabf
from scipy.optimize import curve_fit
from scipy.ndimage import gaussian_filter1d
```

```python
base_folder_path = Path(f'/mnt/isilon/marsh_single_unit/FieldPotentialAnalysis')
abf_folder = "abfs"

abf_folder_path = base_folder_path / abf_folder
abf_subfolder_path = abf_folder_path / abf_subfolder
abf_subfolder_list = glob.glob(f'{abf_folder_path}/*')

print(abf_folder_path)
print(abf_subfolder_path)
```

```python
baseline_window = [0.01, 0.02] # seconds
curvefit_window = [0.00375, 0.02] # seconds

fitting_negslope = -5000

fitting_init = [fitting_negslope, -curvefit_window[0] * fitting_negslope, 100, 3000, 0.005]
fitting_bounds = ([-np.inf, 0, 1, 1000, 0.002],
                  [-500, np.inf, np.inf, np.inf, 0.006])
def spike_fitting_function(t, m, b, decay, sig_sharp, sig_offset):
    sig = 1 / (1 + np.exp(-sig_sharp * (t - sig_offset)))
    return -np.exp(-decay * t) * sig + (m*t + b) * (1 - sig)

# fitting_init = [5000, 0.005, 1000, 0.01, 0]
# fitting_bounds = ([1000, 0, 100, 0, -1],
#                   [np.inf, 0.5, np.inf, 0.5, 1])
def spike_exp_fitting_function(t, d_1, off_1, d_2, off_2, b):
    return np.exp(-d_1 * (t - off_1)) - np.exp(-d_2 * (t - off_2)) + b

fit_function = spike_fitting_function
```

```python
def get_Ycrossings_idx(y, value):
    return np.argwhere(np.diff(np.sign(value - y))).flatten()

def find_slopes_fitting(V, t, f_s, percentile=[0.3, 0.8]):
    popt, _ = curve_fit(spike_fitting_function, t, V, fitting_init, bounds=fitting_bounds)
    fit_V = []
    for t_i in np.arange(0, 1, 1/f_s):
        fit_V.append(spike_fitting_function(t_i, *popt))
    fit_V = np.array(fit_V)
    thresh_low = np.min(fit_V) * percentile[0]
    thresh_high = np.min(fit_V) * percentile[1]
    
    idx_low = get_Ycrossings_idx(fit_V, thresh_low)
    idx_high = get_Ycrossings_idx(fit_V, thresh_high)
    
    decayslope = (fit_V[idx_low[-1]] - fit_V[idx_high[-1]]) / ((idx_low[-1] - idx_high[-1]) / f_s)
    riseslope = (fit_V[idx_low[0]] - fit_V[idx_high[0]]) / ((idx_low[0] - idx_high[0]) / f_s)
    
    return (riseslope, decayslope), popt
    # return (popt[0], decayslope), popt

def find_slopes_traditional(V, t, f_s, percentile=[0.3, 0.8], baseline=None, smoothing=None):
    idx_peak = np.argmin(V)
    V_peak = V[idx_peak]
    
    if smoothing is not None:
        V = gaussian_filter1d(V, smoothing)
    
    if baseline is not None:
        V_base = baseline
    else:
        V_base = np.max(V[idx_peak:])
    
    thresh_low = V_base + (V_peak - V_base) * percentile[0]
    thresh_high = V_base + (V_peak - V_base) * percentile[1]
    
    
    idx_low = get_Ycrossings_idx(V[idx_peak:], thresh_low)
    idx_high = get_Ycrossings_idx(V[idx_peak:], thresh_high)

    if not idx_low.any():
        print("idx_low is empty! Using max value after spike")
        idx_low = [np.argmax(V[idx_peak:])]

    if not idx_high.any():
        print("idx_high is empty! Using min value after spike")
        idx_high = [np.argmin(V[idx_peak:])]

    decayslope = (V[idx_peak + idx_low[0]] - V[idx_peak + idx_high[0]]) / ((idx_low[0] - idx_high[0]) / f_s)

    idx_low = get_Ycrossings_idx(V[:idx_peak], thresh_low)
    idx_high = get_Ycrossings_idx(V[:idx_peak], thresh_high)
    
    if not idx_low.any():
        print("idx_low is empty! Using max value before spike")
        idx_low = [np.argmax(V[:idx_peak])]

    if not idx_high.any():
        print("idx_high is empty! Using min value before spike")
        idx_high = [np.argmin(V[:idx_peak])]
    
    riseslope = (V[idx_low[-1]] - V[idx_high[-1]]) / ((idx_low[-1] - idx_high[-1]) / f_s)

    return (riseslope, decayslope), V
```

```python
abf_file_list = glob.glob(f'{abf_subfolder_path}/*.abf')
abf_file_list.sort()

baseline_window = [0.01, 0.025] # seconds
curvefit_window = [0.00375, 0.025] # seconds
```

```python
rise_slope_fit = []
rise_slope_trad = []
decay_slope_fit = []
decay_slope_trad = []

for abf_file_path in abf_file_list:
    
    abf = pyabf.ABF(abf_file_path)
    units = abf.adcUnits
    f_s = abf.sampleRate

    print(units)
    # print(abf.headerText)

    # Extract V and t data from .ABF
    sweep_V = []
    sweep_t = []
    for i in range(abf.sweepCount):
        abf.setSweep(i)
        sweep_V.append(abf.sweepY)
        sweep_t = abf.sweepX
        
    sweep_V = np.array(sweep_V)
    sweep_t = np.array(sweep_t)

    # Calculate average and max for baselines
    sweep_V = np.average(sweep_V, axis=0)
    sweep_at_baseline = sweep_V[round(baseline_window[0] * f_s) : round(baseline_window[1] * f_s)]
    baseline_V = np.average(sweep_at_baseline)
    basemax_V = np.max(sweep_at_baseline)


    curvefit_window_samples = [round(e * f_s) for e in curvefit_window]
    fitting_t = sweep_t[curvefit_window_samples[0] : curvefit_window_samples[1]]
    fitting_V = sweep_V[curvefit_window_samples[0] : curvefit_window_samples[1]]
    # print(fitting_t.shape)
    # print(fitting_V.shape)

    (m_rise, m_decay), popt = find_slopes_fitting(fitting_V, fitting_t, f_s)
    (m_riset, m_decayt), V_smooth = find_slopes_traditional(fitting_V, fitting_t, f_s, 
                                                            baseline=None,
                                                            smoothing=5)
    print(m_rise, m_decay)
    print(m_riset, m_decayt)
    
    rise_slope_fit.append(m_rise)
    decay_slope_fit.append(m_decay)
    rise_slope_trad.append(m_riset)
    decay_slope_trad.append(m_decayt)

    # popt, _ = curve_fit(fit_function, fitting_t, fitting_V, fitting_init, bounds=fitting_bounds)
    
    fitted_V = []
    fitinit_V = []
    for fit_t in sweep_t:
        fitted_V.append(fit_function(fit_t, *popt))
        fitinit_V.append(fit_function(fit_t, *fitting_init))
    fitted_V = np.array(fitted_V)
    fitinit_V = np.array(fitinit_V)
    
    # Plot figures
    fig, ax = plt.subplots(1, 1, figsize=(7, 2))
    ax.set_ylim(-6, 1)
    ax.set_xlim(0, curvefit_window[1])

    ax.plot(sweep_t, sweep_V)
    ax.plot(sweep_t, fitted_V)
    # ax.plot(sweep_t, fitinit_V)
    ax.plot(fitting_t, V_smooth, ls='--', lw=2, alpha=0.3)
    
    ax.hlines(baseline_V, baseline_window[0], baseline_window[1], color='C2')
    ax.hlines(basemax_V, baseline_window[0], baseline_window[1], color='C3')
    ax.axvspan(curvefit_window[0], curvefit_window[1], alpha=0.2, color='yellow')
    # plt.plot(sweep_t, sweep_C)

    ax.set_ylabel("Voltage (mV)")
    ax.set_xlabel("Time (s)")    
    plt.show()
    
```
